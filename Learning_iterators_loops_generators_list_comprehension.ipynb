```python
my_list = [6, 8, 1, 4, 10, 7, 8, 9, 3, 2, 5]
my_dict = {'py': 'python', 'rb': 'ruby', 'js': 'javascript'}

# sum of all ints
sum = 0

for num in my_list:
    sum += num
print(f"Sum using list: {sum}")

print()
sum = 0

for num in range(11):
    sum += my_list[num]
print(f"Sum using range generator: {sum}")

print()
sum = 0

for num in range(len(my_list)):
    sum += my_list[num]
print(f"Sum using range generator with range auto-calculation: {sum}")

print()
run_times = int(input('How many times do you want to run? '))

for num in range(run_times):
    print(f"Run: {num + 1}")
    
print()
# Print all even numbers between 6 - 20
for i in range(6, 21, 2):
    print(i, end=' ')
print()

# Loop over my_dict
for item in my_dict:
    print(item)
    
for key, value in my_dict.items():
    print(f"Key is {key}, Value is {value}")


# Loop over my_dict and list keys
for key in my_dict.keys():
    print(f"Key: {key}")
    
print()

# List Comprehension

from random import randint
# generate 100 random integers between 1 and 100
print([randint(1, 100) for x in range(101)])

print()
# display the squared value for each number between 1 - 10
print([num**2 for num in range(11)])

print()
from random import randint, choice
from string import ascii_lowercase

# generate a list of 100 random ASCII characters 
chars = [choice(ascii_lowercase) for num in range(100)]
print(chars)

# builde a list of all the vowels found in chars (duplicates included)
VOWELS = ['a', 'e', 'i', 'o', 'u']
hits = [x for x in chars if x in VOWELS]
for h in hits:
    print(h, end=' ')
print()

# NOTE: set is unordered
for i in set(hits):
    print(i, end=' ')
print()


```
